```python
import numpy as np
import cv2

img = cv2.imread("lena.jpg")

cv2.imshow(img)
```

```python
from google.colab.patches import cv2_imshow

img = cv2.imread("lena.jpg")
img2 = cv2.imread("prueba.jpg")

cv2_imshow(img)
cv2_imshow(img2)
```

```python
imagen=np.array(img)
imagen2=np.array(img2)

imagen3=imagen+imagen2

cv2_imshow(imagen)
```

```python
imagen=np.array(img)
imagen2 = np.array(img2).astype(np.uint8)

alto = 630
ancho = 630
medidas = (alto, ancho)

imagen2=cv2.resize(imagen2, medidas, interpolation=cv2.INTER_LINEAR)

imagen3=imagen-imagen2

cv2_imshow(imagen3)
```
