```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2. imread('silueta.jpg',0)
```

```python
ret, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(img)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))
```

Esqueleto

```python
preimagen= cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=2)
preimagen= cv2.morphologyEx(preimagen, cv2.MORPH_CLOSE, kernel, iterations=2)
cv2_imshow(preimagen)
erosion = cv2.erode(img, kernel, iterations=1)
```

```python
apertura= cv2.morphologyEx(preimagen, cv2.MORPH_OPEN, kernel, iterations=1)
erosion = cv2.dilate(esqueleto, kernel, iterations=1)
esqueleto=  erosion - apertura 
imagenes= np.hstack((img, erosion, apertura, esqueleto))
cv2_imshow(imagenes)
```
