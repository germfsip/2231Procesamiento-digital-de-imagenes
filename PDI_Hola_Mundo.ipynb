```python
import matplotlib.pyplot as plt
from PIL import Image, ImageChops, ImageEnhance, ImageOps
import numpy as np

imagen = Image.open("lena.jpg")
#print(imagen)


ima = np.array(imagen)
print(ima)
plt.imshow(ima)
```

```python
nueva = ImageChops.invert(imagen)
nuevaima=np.array(nueva)
print(nuevaima)
plt.imshow(nuevaima)
```
